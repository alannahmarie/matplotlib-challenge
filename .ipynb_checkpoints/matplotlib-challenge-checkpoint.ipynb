## Observations and Insights 

+ The data provided included duplicate timepoints for mouse g989. All data for this mouse had to be excluded from the analysis. Though the timepoints were simply duplicated, other data such as Tumor Volume and Metastic Sites differed. There was no way to tell which timepoints were the accurate data. 

+ Both male and female mice were fairly evenly represented within the study. 

+ As both the scatter plot and line regression model indicate, there is a positive correlation between mouse weight and tumor volume. 

+ For mouse s185 it appears that the Capsomulin drug regimen worked to diminish the tumor volume over time. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_metadata, study_results, how="left", on="Mouse ID")

# Display the data table for preview
combined_data_df.set_index('Mouse ID')
```

```python
# Checking the number of mice.
number_of_mice = combined_data_df.loc[:,"Mouse ID"].unique()
len(number_of_mice)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = combined_data_df.duplicated(subset=["Mouse ID", "Timepoint"], keep=False).value_counts()
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_data_df.loc[combined_data_df.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data_df = combined_data_df.loc[combined_data_df["Mouse ID"] != "g989"]
```

```python
clean_data_df.set_index("Mouse ID")
```

```python
# Checking the number of mice in the clean DataFrame.
number_of_mice_clean = clean_data_df.loc[:,"Mouse ID"].unique()
len(number_of_mice_clean)
```


## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drug_tumor_table = clean_data_df[["Drug Regimen", "Tumor Volume (mm3)"]]

mean = drug_tumor_table.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median = drug_tumor_table.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variance = drug_tumor_table.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
std_dev = drug_tumor_table.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
std_error = drug_tumor_table.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

summary_table = pd.DataFrame({"mean": mean,
                                "median": median,
                                "var": variance, 
                                "std": std_dev,
                                "sem": std_error})
summary_table.style.set_caption("Tumor Volume(mm3)")
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

aggregate_summary_table = drug_tumor_table.groupby('Drug Regimen').aggregate(['mean', np.median, np.var, np.std, 'sem'])

aggregate_summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
quartile_df = clean_data_df.set_index("Drug Regimen")
quartile_df = quartile_df.groupby("Drug Regimen")["Timepoint"].max()
final_quartile_df = pd.merge(clean_data_df, quartile_df, on=["Drug Regimen","Timepoint"])
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
volume_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 


for drug in treatments:
    
    quartile_volume = final_quartile_df.loc[final_quartile_df["Drug Regimen"] == drug]["Tumor Volume (mm3)"]
    volume_data.append(quartile_volume)
    
quartiles = quartile_volume.quantile([0.25,0.5,0.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

print(f"The lower quartile of the final tumor volume is: {round(lowerq)}")
print(f"The upper quartile of the final tumor volume is: {round(upperq)}")
print(f"The interquartile range of the final tumor volume is: {round(iqr)}")
print(f"The the median of the final tumor volume is: {round(quartiles[0.5])}")
print("-------------------------------------------------------------------------")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.figure(figsize=(16,8))
plt.boxplot(volume_data)
plt.title("Tumor Volume Across 4 Regimens: \n Capomulin, Ramicane, Infubinol, and Ceftamin \n", fontsize = 15)
plt.xlabel("Drug Regimen\n ", fontsize = 13)
plt.ylabel("Tumor Volume (mm3)\n ", fontsize = 13)
plt.xticks([1, 2, 3, 4], treatments)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
time_tumor_table = clean_data_df[["Tumor Volume (mm3)", "Timepoint", "Drug Regimen", "Mouse ID"]]
time_tumor_table = time_tumor_table[time_tumor_table["Drug Regimen"] == "Capomulin"]
time_tumor_table = time_tumor_table.loc[time_tumor_table["Mouse ID"] == "s185"]

plt.figure(figsize=(16,8))
plt.plot(time_tumor_table["Timepoint"], time_tumor_table["Tumor Volume (mm3)"], marker="8", color="maroon")
plt.title("Mouse ID # s185 \n Tumor Volume vs. Timepoint\n ", fontsize=14)
plt.xlabel("Timepoint", fontsize=13)
plt.ylabel("Tumor Volume  (mm3)", fontsize=13)
plt.grid()
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
tumor_weight_table = clean_data_df[["Tumor Volume (mm3)", "Weight (g)", "Drug Regimen", "Mouse ID"]]
cap_tumor_weight_table = tumor_weight_table[tumor_weight_table["Drug Regimen"] == "Capomulin"]
cap_tumor_weight_table.set_index("Drug Regimen", inplace=True)

cap_mouse_weight = cap_tumor_weight_table["Weight (g)"]
cap_tumor_vol = cap_tumor_weight_table["Tumor Volume (mm3)"]
plt.figure(figsize=(16,8))
plt.scatter(cap_mouse_weight, cap_tumor_vol)
plt.xlabel('Mouse Weight (Grams)', fontsize=13)
plt.ylabel('Tumor Volume (mm3)', fontsize=13)
plt.title("Tumor Volume vs. Mouse Weight\n ", fontsize=14)
plt.grid()
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
avg_cap_tumor_vol = cap_tumor_weight_table.groupby("Weight (g)").mean()
avg_cap_tumor_vol.reset_index(inplace=True)
mouse_weight = avg_cap_tumor_vol["Weight (g)"]
avg_tumor_vol = avg_cap_tumor_vol["Tumor Volume (mm3)"]

correlation = st.pearsonr(mouse_weight, avg_tumor_vol)
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}.")
```

```python
x_values = mouse_weight
y_values = avg_tumor_vol

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(12, 8))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(21,40),fontsize=18,color="red")
plt.xlabel(" \n Mouse Weight (Grams)", fontsize=13)
plt.ylabel("Tumor Volume (mm3) \n ", fontsize=13)
plt.title(" \n Linear Regression Model \n Mouse Weight and Average Tumor Volume\n ", fontsize=14)
plt.tight_layout()
plt.show()
```
